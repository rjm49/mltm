```python
import numpy
import numpy as np

item_w = 3
n_items_in_bank = 100
n_users = 1000
av_pass_rate = 0.5
items_are_conj = True

def sigmoid(z):
    return 1/(1 + np.exp(-z)) 

def inv_sigmoid(p):
    return -np.log((1/p)-1)

ptest=sigmoid(3)
ztest=inv_sigmoid(ptest)
assert np.isclose(0.95257, ptest)
assert np.isclose(3, ztest)

print("..." , inv_sigmoid(0.5))

av_KCp = av_pass_rate**(1/item_w) if items_are_conj else 1/item_w
av_KCs = inv_sigmoid(av_KCp)
print(av_KCs)

users = np.random.normal(size=(n_users, item_w), loc=av_KCs, scale=0.5)
items = np.random.normal(size=(n_items_in_bank, item_w), loc=0, scale=0.5)
print(users)
print(items)

ps = []
for s in users:
    for q in items:
        logit_differences = s-q
        if items_are_conj:
            p = np.prod(sigmoid(logit_differences))
        else:
            p = sigmoid(np.sum(logit_differences))
        ps.append(p)
print(numpy.mean(ps), numpy.median(ps))
```
